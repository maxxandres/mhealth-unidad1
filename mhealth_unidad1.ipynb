```python
import glob

# Buscar todos los .log dentro de la carpeta
archivos = glob.glob("mhealth_data/MHEALTHDATASET/*.log")

# Leer y concatenar
dfs = []
for archivo in archivos:
    df_temp = pd.read_csv(archivo, sep="\t", header=None)
    # agregar columna con el nombre del archivo
    df_temp["Subject"] = archivo.split("/")[-1].replace(".log", "")
    dfs.append(df_temp)

df_total = pd.concat(dfs, ignore_index=True)

print("Dimensiones del dataset completo:", df_total.shape)
df_total.head()
```

```python
# Información general
df_total.info()

```

### Información general del dataset

- El dataset consolidado contiene **1.217.545 filas** y **25 columnas**.  
- La mayoría de las variables son de tipo `float64` (23 columnas), lo que corresponde a las señales de los distintos sensores (, ECG, etc.).  
- La penúltima columna corresponde a la **etiqueta de actividad** (`int64`), que indica qué actividad estaba realizando el sujeto en ese instante.  
- Se añadió una columna adicional llamada **`subject`** (`object`), la cual identifica a qué archivo/log pertenece cada observación (es decir, a qué voluntario del experimento corresponde).  
- No se detectan valores faltantes (`Non-Null Count` es igual al total de filas en todas las columnas).  

**Conclusión inicial:** el dataset está completo, es grande (más de un millón de registros), y las variables ya están en formato numérico adecuado para su análisis.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_scaled = df_total.copy()

# Escalar todas las columnas menos la etiqueta y el subject
cols_sensores = df_total.columns[:-2]
df_scaled[cols_sensores] = scaler.fit_transform(df_total[cols_sensores])

df_scaled.head()
```

### Normalización / Estandarización

- Se aplicó la técnica de **StandardScaler** de *scikit-learn* a todas las variables de sensores.  
- Este procedimiento transforma cada variable para que tenga:
  - Media = 0  
  - Desviación estándar = 1  
- Con esto se eliminan las diferencias de escala entre señales que originalmente estaban en unidades distintas (por ejemplo, acelerómetro en "g", ECG en milivoltios, etc.).  
- La estandarización es fundamental para que en etapas futuras los algoritmos de machine learning no se vean dominados por variables con valores numéricamente más grandes.
